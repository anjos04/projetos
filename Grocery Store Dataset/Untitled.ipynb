```python
import pandas as pd
df = pd.read_csv("GroceryDataset.csv")
df
```

```python
df.info()
```

```python
df["Price"] = df["Price"].apply(lambda x: str(x).replace(',','.'))
```

```python
df['Price'] = df['Price'].str.strip('$')
df
#dflimpo = df['Price'].str.strip('$')
#dflimpo
```

```python
df['Price'] = df['Price'].str.replace(r'\D', '')
```

```python
df['Price'] = df['Price'].str.strip()  # Remove leading/trailing spaces
```

```python
df['Price'] = df['Price'].astype('float64')
```
